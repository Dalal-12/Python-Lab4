```python
while True:
    try:
        name=input("Enter your name : ")
        age=int(input("Enter your age : "))
        if age in range(0,6):
            print("Kid")
        elif age < 0 :
            print("Wrong age value, you can not enter value 0 or less")
        else:
            total_cost=0
            while True:
                try:
                    num_of_items=int(input("Enter the number of item you want to buy : "))
                except:
                    print("Enter a valid number of item")
                else:
                    break
            x=1
            while x <= num_of_items :
                while True:
                    try:
                        item_name=input("Enter the Item name : ")
                    except:
                        print("Enter a Correct Name Please .. ")
                    else:
                        break
                while True:
                    try:
                        item_price=int(input("Enter the item price : "))
                    except:
                        print("Enter a Valid Price Value Please .. ")
                    else:
                         break
                while True:
                    try:
                        item_quantity=int(input("Enter the item quantity : "))
                    except:
                        print("Enter a Valid Quantity Value Please .. ")
                    else:
                        break
                item_cost=item_price*item_quantity
                total_cost += item_cost
                print(f'this item costs: {item_cost} SR')
                print("-"*30)
                x += 1
            print(f'The Total Cost : {total_cost} SR')
            if total_cost > 200:
                total_cost=total_cost - (total_cost/100)*20
                print(f'The Total Cost After Discount : {total_cost} SR')
    except:
        print('Enter a Valid Information Please')
    else:
        repeat=input("Do you want to repeat the process ? Type Yes or No : ")
        if repeat == "No" or repeat=="no" :
            break

```
